Difference between Simple Linear Regression (SLR) and Multiple Linear Regression (MLR):

SLR models the relationship between one predictor (independent variable) and one outcome (dependent variable), fitting a straight line to predict the outcome.
MLR extends this by incorporating two or more predictors. The benefit of MLR is that it allows for the inclusion of multiple factors that may influence the outcome, improving the model’s explanatory power and providing a more nuanced understanding of relationships between variables.
Difference between a continuous variable and an indicator variable in Simple Linear Regression:

A continuous variable can take on any numeric value within a range and represents a gradual change in the outcome. The regression line will show how the outcome varies continuously with changes in the predictor.
An indicator variable (also called a dummy variable) represents categories (often 0 or 1), where the outcome changes discretely between groups, such as male/female or before/after.
These two forms:

Continuous predictor: Y = β₀ + β₁X (where X is continuous).
Indicator predictor: Y = β₀ + β₁D (where D is 0 or 1).
Change in model behavior with the introduction of an indicator variable in Multiple Linear Regression:

When an indicator variable is added alongside a continuous variable, the model adjusts to account for shifts in the intercept between groups (represented by the indicator) while still allowing the continuous variable to influence the outcome within each group. This results in parallel lines for each group, where the continuous variable’s effect is maintained but starting points (intercepts) differ.
SLR form: Y = β₀ + β₁X MLR form: Y = β₀ + β₁X + β₂D

Effect of adding an interaction between a continuous and an indicator variable in MLR:

When an interaction term between a continuous variable and an indicator variable is added, the model allows the slope of the continuous variable to differ across the categories of the indicator variable. This creates different rates of change (slopes) for each group, rather than just different intercepts.
MLR with interaction: Y = β₀ + β₁X + β₂D + β₃(X * D) This allows both intercept and slope to vary by group.

Behavior of MLR model with indicator variables from a non-binary categorical variable:

When a categorical variable with more than two levels is encoded as indicator variables, MLR assigns a separate binary variable for each level (excluding a reference level). Each indicator captures the shift in the outcome for that category compared to the reference. The model behaves by adjusting the intercept for each category but assumes the same slope for continuous predictors unless interactions are added.
The form is: Y = β₀ + β₁X + β₂D₁ + β₃D₂ + ... + βₙDₙ

This results in binary encodings (e.g., 1 for belonging to a category, 0 otherwise) and allows the model to compare groups systematically.

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Placeholder for your dataset (replace 'pokeaman' with your actual dataset)
# pokeaman = pd.read_csv('your_dataset.csv')

# Define the model specification (adjust based on your specific dataset and columns)
model_formula = 'HP ~ Attack + Defense + Speed + Legendary + Q("Sp. Def") + Q("Sp. Atk") + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'

# Initialize lists to store performance metrics
in_sample_r2 = []
out_of_sample_r2 = []

# Number of iterations for random splits and model fitting
n_iterations = 50

# For loop to repeat the process multiple times
for i in range(n_iterations):
    # Split the data randomly into train/test (without fixed random seed)
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # Fit the model to the training data
    model_spec = smf.ols(formula=model_formula, data=pokeaman_train)
    model_fit = model_spec.fit()
    
    # Get "in-sample" R-squared (from the training data)
    in_sample_r2.append(model_fit.rsquared)
    
    # Get predicted values for "out-of-sample" (test data)
    yhat_out_of_sample = model_fit.predict(pokeaman_test)
    y_out_of_sample = pokeaman_test['HP']
    
    # Calculate "out-of-sample" R-squared using the actual test set vs predicted values
    out_of_sample_r2.append(np.corrcoef(y_out_of_sample, yhat_out_of_sample)[0, 1]**2)

# Visualize the results
plt.figure(figsize=(10, 6))

# Plot the in-sample and out-of-sample R-squared values
plt.plot(range(n_iterations), in_sample_r2, label='In-Sample R-squared', marker='o')
plt.plot(range(n_iterations), out_of_sample_r2, label='Out-of-Sample R-squared', marker='x')

# Add labels and title
plt.xlabel('Iteration')
plt.ylabel('R-squared')
plt.title('In-Sample vs Out-of-Sample Model Performance (Multiple Iterations)')
plt.legend()

# Show the plot
plt.show()

# Optionally, print out some summary statistics of the R-squared values
print(f'Average in-sample R-squared: {np.mean(in_sample_r2):.4f}')
print(f'Average out-of-sample R-squared: {np.mean(out_of_sample_r2):.4f}')
````
